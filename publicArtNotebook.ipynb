# Public Art Submetric

This notebook ranks the neighborhoods by how many public art installations it has


**Importing**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('publicArtData.csv', sep=',')

trashCount = data.groupby('neighborhood')['id'].count()
trashCount = trashCount.sort_values()
```

**Plot**

```python
plt.figure(figsize=(12, 8))
trashCount.sort_values().plot(kind='bar', color='pink')
plt.title('Amount of public art installations per neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('# of public art installations')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
